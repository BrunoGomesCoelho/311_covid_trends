```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv'
df_all = pd.read_csv(url, low_memory=False)
```

```python
df_all.head()
```

```python
df_nyc = df_all[(df_all['county'] == 'New York City') & (df_all['date'] <= '2020-10-31')].copy()
```

```python
df_nyc.rename(columns={'date': 'day'}, inplace=True)
```

```python
df_nyc['new_cases'] = df_nyc['cases'].diff()
df_nyc['new_deaths'] = df_nyc['deaths'].diff()
```

```python
df_nyc['week_avg_new_cases'] = df_nyc['new_cases'].rolling(7).mean()
df_nyc['week_avg_new_deaths'] = df_nyc['new_deaths'].rolling(7).mean()
```

```python
df_nyc['deaths'] = df_nyc['deaths'].astype('Int64')
df_nyc['cases'] = df_nyc['cases'].astype('Int64')
df_nyc['new_deaths'] = df_nyc['new_deaths'].astype('Int64')
df_nyc['new_cases'] = df_nyc['new_cases'].astype('Int64')
```

```python
df_nyc
```

```python
df_nyc.to_csv('covid.csv',
              columns=['day', 'cases', 'deaths',
                     'new_cases', 'new_deaths',
                     'week_avg_new_cases',
                     'week_avg_new_deaths'],
              index=False)
```
